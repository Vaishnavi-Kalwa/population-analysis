```python
country=data['Continent'].value_counts()
print(country)
```

```python

plt.figure(figsize=(6,4))
plt.bar(country.index,country.values, color='Purple')
plt.xlabel('Continent')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45)
plt.title('Number of Countries in each Continent')
plt.show()
```

```python
unique_countries = data['Country/Territory'].unique()
num_countries = len(unique_countries)
cmap = plt.cm.get_cmap('viridis', num_countries)
colors = [cmap(i) for i in range(num_countries)]
plt.figure(figsize=(10, 6))
for i, country in enumerate(unique_countries):
    country_data = data[data['Country/Territory'] == country]
    plt.bar(country_data.index, country_data['2022 Population'], color=colors[i], label=country)

plt.xlabel('Index')
plt.ylabel('2022 Population')
plt.title('2022 Population by Country/Territory')
plt.legend(bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
num_feat=data.select_dtypes(include='number').columns
continent=data.groupby('Continent')[num_feat].mean()
population_features = ['2022 Population','2020 Population','2015 Population','2010 Population','2000 Population','1990 Population','1980 Population','1970 Population']

plt.figure(figsize=(10,5))
for feature in population_features:
    plt.plot(continent[feature],label = feature)
plt.legend()
plt.title('Year-to-year population change of continents', size=24);
plt.show()
```

```python

top_countries = data.groupby(by = 'Country/Territory')['2022 Population'].sum().sort_values(ascending=False).head(10)

fig=px.bar(x=top_countries.index,y=top_countries.values,color=top_countries.index,
          color_discrete_sequence=px.colors.sequential.PuRd,
          text=top_countries.values,
          title="Top 10 Countries with the most population",
          template= 'plotly_dark')
fig.update_layout(  xaxis_title="Countries", yaxis_title="Population", autosize=False, width=500, height=400)
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.show()
```

```python
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x=data[['Area (km²)','Density (per km²)','Growth Rate']]
y=data['World Population Percentage']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.7,train_size=0.3,random_state=42,shuffle=False)
```

```python
x_train_constant=sm.add_constant(x_train)
model=sm.OLS(y_train, x_train_constant).fit()
print(model.summary())
```

```python
x1=data.iloc[:,6:13]
y1=data['2020 Population']
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.8,train_size=0.2,random_state=42,shuffle=False)
x1_train_constant=sm.add_constant(x1_train)
model=sm.OLS(y1_train, x1_train_constant).fit()
print(model.summary())
```

```python

lm=LinearRegression()
lm.fit(x1_train,y1_train)
y1_pred=lm.predict(x1_test)
lm_pred=pd.DataFrame({'Predicted':y1_pred,'Actual':y1_test})
print(lm_pred)

print("MSE:",np.sqrt(mean_squared_error(y1_test,y1_pred)))
print("MAE:",mean_absolute_error(y1_test,y1_pred))

print("R2:",r2_score(y1_test,y1_pred))
```

```python

from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor(n_neighbors=3)
knn.fit(x1_train,y1_train)
y1_pred_k=knn.predict(x1_test)
knn_pred=pd.DataFrame({'Predicted':y1_pred_k,'Actual':y1_test})
print(knn_pred)

print("MSE:",np.sqrt(mean_squared_error(y1_test,y1_pred_k)))
print("MAE:",mean_absolute_error(y1_test,y1_pred_k))
print("R2:",r2_score(y1_test,y1_pred_k))

```

```python

from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(random_state=42)
dt.fit(x1_train,y1_train)
y1_pred_d=dt.predict(x1_test)
dt_pred=pd.DataFrame({'Predicted':y1_pred_d,'Actual':y1_test})
print(dt_pred)

print("MSE:",np.sqrt(mean_squared_error(y1_test,y1_pred_d)))
print("MAE:",mean_absolute_error(y1_test,y1_pred_d))
print("R2:",r2_score(y1_test,y1_pred_d))
```

```python

from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
rf.fit(x1_train,y1_train)
y1_pred_r=rf.predict(x1_test)
r_pred=pd.DataFrame({'Predicted':y1_pred_r,'Actual':y1_test})
print(r_pred)

print("MSE:",np.sqrt(mean_squared_error(y1_test,y1_pred_r)))
print("MAE:",mean_absolute_error(y1_test,y1_pred_r))
print("R2:",r2_score(y1_test,y1_pred_r))
```
